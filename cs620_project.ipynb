# **Introduction ‚úÖ**
-------

Richard Dawkins, a famous evolutionary biologist, coined the term "meme" denoting a gene of information existing in the pool of ideas: "Just as genes propagate themselves in the gene pool by leaping from body to body via sperms or eggs, so memes propagate themselves in the meme pool by leaping from brain to brain" (Dawkins 1976).

Anything can be a meme: a rumor, a funny picture, a political ideology etc. Similarly to biological genes, not all memes survive the process of natural selection, and it is important for them to have a set of characteristics that fit their environment.

In this work, we will be studying a particular subset of memes - emoji - that are, like words, are used to transmit ideas and moods in an environment of the social media. We will use Twitter as an environment and tweets as carriers of emojies (just like our bodies carry the genes). 

Emojis can perform variable functions in communication: they can either be used instead of nouns or verbs or adjectives substituting those in otherwise grammatically complete sentences, or enrich such sentences via additional emotional meaning (Na‚Äôaman et al. 2017). Also, the meaning of an emoji might vary among societies, age, and social groups, and be perceived differently (Barbieri et al. 2016). It has been shown that patterns of emoji usage vary from country to country and the distribution of different emoji usage is similar to one that can be found in biological communities of organisms (Kerjiwal et al. 2021).

There will be a set of metrics we will be using to quantify the ability of memes to succesfully reproduce and outcompetes other ideas:

*   Tweet-specific metrics:

> number of likes

> number of comments

> number of retweets (all normalized by the number of followers of the account)

> overall number of emoji used

> emoji richness (number of unique emojis)

> emoji per word

> emoji per character

> amount of redindant emoji (repeat themselves or respective word)

*   Emoji-specfic metrics

> type

> sentimental characteristics

> year when introduced


Using these characteristics, we will seek to answer several questions:

* what combinations and amounts of emojis are related to the strongest tweet success (it means that we also will need tweets with no emoji)?

* what intrinsic characteristics of emoji make them competitive?

* is there a relationship between the use of emoji in advertisement campaigns and commercial profitability?

## **Obtaining the Twitter1 dataset** ‚úÖ

The Twitter1 dataset is supposed to contain a collection of tweets from any year tweeted by users that have more than 1000 followers. This way, we ensure that we do not account for tweets that are invisible to the competition due to limited circle of users that can see and like (or unlike) the idea of a tweet.

The tweets were obtained using [Twitter academic access API](https://developer.twitter.com/en/products/twitter-api/academic-research). We limited our queries to tweets in English and those that contained at least one letter of the English alphabet. To avoid possible bias caused by the dynamics of Twitter popularity among users, we stratified the search by each calendar day between 01 Jan 2007 (as the social media was launched in July of 2006) and 26 Sep 2022.

The following variables were included: author ID, date when the tweet was created, conversation ID, tweet ID, language, count of retweets, replies, likes, and quotes, tweet source (i.e., using which platform the tweet was published), text of the tweet, date when the author's profile was created, username, whether an account is verified by Twitter, ccount of followers, following, and tweets published by the account.

We developed the methods of this work on a smaller dataset containing 100 tweets per day, so that the code could be applied to a big dataset containing roughly 1,000 tweets per day, resulting in a dataset containing up to 2.5 millions unique tweet records.

We acknowledge the fact that a major proportion of the tweets might be considered "invisible" by the Twitter users, meaning that they would not be exposed to assessment by users and will not experience any selection pressure. We assume that these invisible tweets would most likely be published by accounts with a low count of followers, therefore, we (a) show that the tweet metrics are a function of the account metrics, and (b) drop all individual tweets published by accounts with less than an arbitrary number of followers (1,000). After accounting for the invisible tweets, the resulting training dataset contained 100,744 unique tweet records.

```python
import pandas as pd
import requests
import datetime
import string
import time
import os


# bearer's token
os.environ['TOKEN'] = 'Bearer token'

def auth():
    return os.getenv('TOKEN')

def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers

def create_params(keyword, start_date, end_date, max_results=10):
    # change params based on the endpoint you are using
    query_params = {
        'query': keyword,
        'start_time': start_date,
        'end_time': end_date,
        'max_results': max_results,
        'expansions': 'author_id,in_reply_to_user_id,geo.place_id',
        'tweet.fields': 'id,text,author_id,in_reply_to_user_id,geo,conversation_id,created_at,lang,public_metrics,referenced_tweets,reply_settings,source',
        'user.fields': 'id,name,username,created_at,description,public_metrics,verified',
        'place.fields': 'full_name,id,country,country_code,geo,name,place_type',
        'next_token': {}}
    return query_params

def connect_to_endpoint(url, headers, params, next_token=None):
    params['next_token'] = next_token  # params object received from create_url function
    while True:
        response = requests.request("GET", url, headers=headers, params=params)
        print("Endpoint Response Code: " + str(response.status_code))
        if response.status_code == 200:
            return response.json()
        else:
            if response.status_code == 429:
                time.sleep(60)
            else:
                raise Exception(response.status_code, response.text)

def write_to_csv(tweets, users, file_name):
    # Create dataframes from tweets and users, and merge them
    tweets = pd.DataFrame(tweets)
    tweets.rename(columns={'author_id': 'user_id'}, inplace=True)
    tweets = pd.concat([tweets.drop('public_metrics', axis=1), pd.DataFrame(tweets['public_metrics'].tolist())], axis=1)
    users = pd.DataFrame(users)
    users.rename(columns={'id': 'user_id'}, inplace=True)
    users = pd.concat([users.drop('public_metrics', axis=1), pd.DataFrame(users['public_metrics'].tolist())], axis=1)
    mrgd_df = tweets.merge(users, how='left', on='user_id')
    mrgd_df.replace(r'\s+', ' ', regex=True, inplace=True)
    mrgd_df.to_csv(file_name, index=False)

bearer_token = auth()
headers = create_headers(bearer_token)

date_list = pd.date_range(datetime.date(2014, 5, 1), datetime.date(2022, 5, 1)).tolist()
keyword = '(' + ' OR '.join(string.ascii_letters) + ') lang:en'
search_url = "https://api.twitter.com/2/tweets/search/all"

total_tweets = 0  # Total number of tweets we collected from the loop
max_count = 100  # Max tweets per time period, later change to ~4000
tweets, users = [], []  # List to collect the retrieved tweets and users info

for i in range(len(date_list) - 1):
    count = 0  # Counting tweets per time period
    max_tweets, n_tweets = min(500, max_count), min(500, max_count)  # Max tweets per request and # of tweets to request
    next_token = None

    start_date = date_list[i].isoformat() + 'Z'
    end_date = date_list[i + 1].isoformat() + 'Z'
    params = create_params(keyword, start_date, end_date, n_tweets)
    json_response = connect_to_endpoint(search_url, headers, params, next_token)
    result_count = json_response['meta']['result_count']

    if result_count:
        # Add the retrieved tweets to the list
        tweets.extend(json_response['data'])
        users.extend(json_response['includes']['users'])
        # Save the token to use for next call
        next_token = json_response['meta']['next_token']

    print("-------------------")
    print("Start Date: ", date_list[i])
    count += result_count
    total_tweets += result_count
    print("# of Tweets added in the call: ", result_count)
    print("-------------------")
    time.sleep(2)

write_to_csv(tweets, users, '/content/drive/MyDrive/cs620/data-100perday2.csv')
print("Total # of Tweets added: ", total_tweets)
```

## **Analysing emoji's evolutionary success** ‚úÖ

Although the idea of memes is directly inspired bu biological genetical systems, the ecosystem of emojis can hardly be described using standard tools of bioinformatics.

In conventional bioinformatics, information is carried on DNA molecules written as a sequence of four possible nucleobases, or "letters", that are translated into the sequence of twenty types of aminoacids within a protein molecule. With that said, a gene - unit of information - is encoded using long sequences of only four lettes, and the survival of an organism is a function of thousands genes within this organism.

In our case, though, organisms are represented by tweets and genes are represented by words. Analysis of speech would be way to complicated, but analysis of emojis might be less effort-intensive and more standardized.

Here is where the discrepancies between the biological systems and our study system begin. First, there are two possible "tweet-genes": conventional words and emojis, and not all tweets contain either of them. Such a situation is hardly imaginable in biological systems, where there a gene is a gene (period).

Second, encoding of information is vastly different. While genes are long (and variable) sequences of A, C, T, or G, memes are single characters, although in some cases can be encoded with combinations of them. E.g.,

All of these reasons imply that although the whole philosophy of this study is based upon the similarity between genes and memes, as well as mechanism of their selection withon respective systems, these systems are exceedingly different to deploy methodology of genetic studies (although they exist) in the study presented.

The cause to do so still persist, so that we would need to apply an independent methodology to this study.

In order to address the objective related to evolutionary success of emoji, one would need a method to quantify this success. In genes, it can be done via comparison of gene sequences to infer the direction of natural selection (see [Murrell et al. 2015](https://doi.org/10.1093/molbev/msv035), [Rizzato et al. 2016](https://doi.org/10.1186/s12859-016-1135-1)). In our case, we would rather infer whether a meme (i.e., emoji) is successful based on its frequency and dynamics within the tweets ecosystem.

Emoji frequency dynamics is described as a time series. Let define an evolutionary successful emoji as one in which increase of frequency is consistently prevalent in time. For example, let 


*   `ds0` represent a neutral time series (i.e., with random dynamics where the next step is determined as the previous one plus a random normally distributed error with a mean of zero),
*   `dsp` represent a positive time series (i.e., with random dynamics where the next step is determined as the previous one plus a random normally distributed error with positive mean), and
* `dsn` represent a positive time series (i.e., with random dynamics where the next step is determined as the previous one plus a random normally distributed error with negative mean).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = [np.random.normal(loc=0.0, scale=1.0)]
b = [np.random.normal(loc=0.0, scale=1.0)]
c = [np.random.normal(loc=0.0, scale=1.0)]
i = [0]

for j in range(100):
  a.append(a[j-1] + np.random.normal(loc=0.0, scale=1.0))
  b.append(b[j-1] + np.random.normal(loc=1.0, scale=1.0))
  c.append(c[j-1] + np.random.normal(loc=-1.0, scale=1.0))
  i.append(i[j-1]+1);

ds0 = pd.DataFrame({"i": i, "ds0": a, "dsp": b, "dsn": c})

fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.plot(ds0["i"], ds0["ds0"])
ax2.plot(ds0["i"], ds0["dsp"])
ax3.plot(ds0["i"], ds0["dsn"])
```

We can expect then that an evolutionary succesful emoji will tend to have a generally high frequency and overally positive rate of growth. Now, the latter have a lag more than 1 step, but let assume that the larger lag will have a lower weight. Therefore, we may attempt to quantify overall evolutionary success as

$w_{0}\sum_{t \forall t \in \{1, 2, \cdots, T\}}{f_{t}} \cdot w_{1} \sum_{t \forall t \in \{1, 2, \cdots, T-1\}}{\left(f_{t+1} - f_{t}\right)} \cdot w_{2} \sum_{t \forall t \in \{1, 2, \cdots, T-2\}}{\left(f_{t+2} - f_{t}\right)} \cdot (\cdots) \cdot w_T\sum_{t \forall t \in \{1\}}{\left(f_{T} - f_{1}\right)}$

where $T$ is overall number of steps within a time series, $f_t$ is a frequency at time $t$, and $w$ is a weight such that $w_t \propto \frac{1}{t}$ and $\sum_{t \forall t \in \{0, 1, 2, \cdots, T\}}{w_t} = 1$. In other words,

$w_{0}\sum_{t \forall t \in \{1, 2, \cdots, T\}}{f_{t}} \cdot \prod_{t \forall t \in \{1, 2, \cdots, T\}}{\left[ w_t\sum_{\tau \forall \tau \in \{1, \cdots, T - t\}}{\left(f_{\tau+t} - f_{\tau}\right)}\right]}$

```python
import pandas as pd

def ds_success(ds):
  ds = [x for x in ds if isinstance(x, float) or isinstance(x, int)]
  df = pd.DataFrame({"f0": ds})
  T = len(df)
  fsum = [sum(ds)]
  w = [None]
  for t in range(1, T):
      df[("f"+str(t))] = df.f0.rolling(1+t).apply(lambda x: x.iloc[t] - x.iloc[0]);
      fsum.append(sum(df[("f"+str(t))].dropna()))
      w.append(1/t)
  w[0] = sum(w[1:])
  w = [round(x/sum(w), 3) for x in w]
  return round(sum(pd.Series(w)*pd.Series(fsum)), 5);

print(ds_success(a))
print(ds_success(b))
print(ds_success(c))
```

It is notable that the suggested metric is close to zero for a non-growing time series, large and positive for a consistently increasing time series, and large and negative for a consistently decreasing time series, denoting neutral, positive, and negative selection, respectively.

```python
import pandas as pd
import requests
import time
import os

# bearer's token
os.environ['TOKEN'] = 'Your token'

def auth():
    return os.getenv('TOKEN')

def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers

def create_params(ids):
    # change params based on the endpoint you are using
    query_params = {
        'ids': ','.join(ids),
        'expansions': 'author_id,in_reply_to_user_id,geo.place_id',
        'tweet.fields': 'id,text,author_id,in_reply_to_user_id,geo,conversation_id,created_at,lang,public_metrics,referenced_tweets,reply_settings,source',
        'user.fields': 'id,name,username,created_at,description,public_metrics,verified',
        'place.fields': 'full_name,id,country,country_code,geo,name,place_type'}
    return query_params

def connect_to_endpoint(url, headers, params, next_token=None):
    params['next_token'] = next_token  # params object received from create_url function
    while True:
        response = requests.request("GET", url, headers=headers, params=params)
        print("Endpoint Response Code: " + str(response.status_code))
        if response.status_code == 200:
            return response.json()
        else:
            if response.status_code == 429:
                time.sleep(60)
            else:
                raise Exception(response.status_code, response.text)

def write_to_csv(tweets, users, tweet_ids, file_name):
    # Create dataframes from tweets and users, and merge them
    tweets = pd.DataFrame(tweets)
    tweets.rename(columns={'author_id': 'user_id'}, inplace=True)
    tweets = pd.concat([tweets.drop('public_metrics', axis=1), pd.DataFrame(tweets['public_metrics'].tolist())], axis=1)
    users = pd.DataFrame(users)
    users.rename(columns={'id': 'user_id'}, inplace=True)
    users = pd.concat([users.drop('public_metrics', axis=1), pd.DataFrame(users['public_metrics'].tolist())], axis=1)
    mrgd_df = tweets.merge(users, how='left', on='user_id')
    mrgd_df.replace(r'\s+', ' ', regex=True, inplace=True)
    spam_mrgd = pd.merge(mrgd_df, tweet_ids)
    spam_mrgd.to_csv(file_name, index=False)

bearer_token = auth()
headers = create_headers(bearer_token)

search_url = "https://api.twitter.com/2/tweets?"

total_tweets = 0  # Total number of tweets we collected from the loop
max_count = 100  # Max tweets per time period, later change to ~4000
tweets, users = [], []  # List to collect the retrieved tweets and users info

# Read tweet ids to a dataframe and extract a list of ids
tweet_ids = pd.read_excel('tweet_ids.xlsx', skiprows=[0, 1, 2, 3])
tweet_ids.columns = ['id', 'label']
tweet_ids['id'] = ['id'].astype('str')
ids = tweet_ids['id'].to_list()

max_count = 100  # Max tweets per request
tweets, users = [], []  # List to collect the retrieved tweets and users info
for i in range(0, 10000, max_count):
    # Change to the endpoint you want to collect data from
    params = create_params(ids[i:i+max_count])
    json_response = connect_to_endpoint(search_url, headers, params)

    # Add the retrieved tweets to the list
    tweets.extend(json_response['data'])
    users.extend(json_response['includes']['users'])
    count = json_response['data'].__len__()
    total_tweets += count
    print("# of Tweets added in the call: ", count)
    print("-------------------")
    time.sleep(2)

write_to_csv(tweets, users, tweet_ids, '/content/drive/MyDrive/cs620/spam_data_id2.csv')
print("Total # of Tweets added: ", total_tweets)
```

###Building classifier models to find the best one

```python
#!pip install lazypredict
import re
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from lazypredict.Supervised import LazyClassifier

# Creation of the spam words array, this will be used to evaluate the "number of spamwords" feature of the tweets
spam_words_arr = ['$$$', '¬£¬£¬£', 'accounts', 'additional', 'bank', 'bonus', 'cash', 'cost', 'credit', 'earn', '$', 'finance', 'advice', 'freedom', 'investment', 'insurance',
                  'decision', 'invoice', '¬£', 'million', 'account', 'potential', 'earnings', 'refund', 'risk', 'save', 'stock', 'thousands', 'trade', 'dollars', 'ad',
                  'amazing', 'bargain', 'offer', 'cheap', 'clearance', 'congratulations', 'dear', 'market', 'marketing', 'delete', 'fantastic', 'free', 'trial', 'gift',
                  'increase', 'incredible', 'sales', 'traffic', 'junk', 'member', 'easy', 'expires', 'extended', 'fast', 'opportunity', 'performance', 'promise', 'sale',
                  'spam', 'special', 'promotion', 'subscribe', 'promo', 'form', 'quick', 'urgent', 'unbeatable', 'unsubscribe', 'visit', 'website', 'win', 'winner', 'boss',
                  'hosting', 'paid', 'business', 'income', 'profit', 'sleep', 'advertisement', 'mlm', '$$$', '@mlm', '100%', 'satisfied', 'cards', 'money', 'guarantee', 'info',
                  'order', 'now!', '18+', 'adult', 'adults', 'over', '18', '21', 'xxx', '50% off', '#1', '4U', 'follow']

COLS = ['retweet_count', 'reply_count', 'like_count', 'quote_count', 'verified', 'followers_count', 'follow_rate',
        'following_count', 'tweet_count', 'listed_count', 'emoji_count', 'name_emoji_count', 'geo', 'description']

# Method used to create the feature set of a given tweet
def extract_tweet_features(row):
    tweet = row['text']
    urls = len(re.findall("(?P<url>https?://[^\s]+)", tweet))
    hashtags = len(re.findall('#', tweet))
    mentions = len(re.findall('@', tweet))
    questions = len(re.findall('\\?', tweet))
    exclamations = len(re.findall('!', tweet))
    periods = len(re.findall('.', tweet))
    dollors = len(re.findall('$', tweet))
    tokens = re.sub(r"[^a-zA-Z0-9]", " ", tweet.lower()).split()
    spamwords = 0
    for token in tokens:
        if token in blacklist or spam_words_arr:
            spamwords += 1
    features = [spamwords, urls, hashtags, mentions, questions, exclamations, periods, dollors, len(tokens)]
    return features

def process_features(row):
    features = list(map(lambda x: np.nansum(x), [row[i] for i in COLS]))
    return features

def calc_bow(data):
    spam_wordset = {}
    normal_wordset = {}
    for _, row in data.iterrows():
        tokens = re.sub(r"[^a-zA-Z0-9]", " ", row['text'].lower()).split()
        for token in tokens:
            if row['label'] == 1:
                try:
                    spam_wordset[token] += 1
                except:
                    spam_wordset[token] = 1
            else:
                try:
                    normal_wordset[token] += 1
                except:
                    normal_wordset[token] = 1
    for key in spam_wordset.keys():
        if key in normal_wordset.keys():
            spam_wordset[key] -= normal_wordset[key]
    blacklist = dict(sorted(spam_wordset.items(), key=lambda x: x[1], reverse=True)[:15])
    return blacklist

# Reading the dataset and expanding some features to it
data = pd.read_csv('/content/drive/MyDrive/cs620/clean_spam_data_emojis.csv')
data['follow_rate'] = data['followers_count'] / data['following_count']
data['follow_rate'] = np.where(data['follow_rate'] == np.inf, data['followers_count'], data['follow_rate'])
data['follow_rate'].fillna(0, inplace=True)
data['description'] = data['description'].apply(lambda x: len(x) if pd.notna(x) else 0)
data['geo'] = data['geo'].apply(lambda x: 1 if pd.notna(x) else 0)
data['referenced_tweets'] = data['referenced_tweets'].apply(
    lambda x: x.split("'")[3] if pd.notna(x) else 'tweet')
data = pd.get_dummies(data, columns=['referenced_tweets'])
blacklist = calc_bow(data)
training_cols = data.columns[26:]
COLS.extend(training_cols)
# Creating the feature set and prepare x and y for the classification
x, y = [], []
for _, row in data.iterrows():
    features = process_features(row)
    tweet_features = extract_tweet_features(row)
    features.extend(tweet_features)
    x.append(features)
    y.append(row['label'])
x_train, x_test, y_train, y_test = \
    train_test_split(np.array(x), np.array(y),
                     stratify=np.array(y),test_size=0.3, random_state=28)
# Using lazy classifier to apply different classifiers
lz_clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
_, scores = lz_clf.fit(x_train, x_test, y_train, y_test)
scores
```

```python
#!pip install wordcloud
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import emojis

class EmojiCloud:
    def __init__(self, font_path='Symbola.otf'):
        self.font_path = font_path
        self.word_cloud = self.initialize_wordcloud()
        self.emoji_probability = None
        
    def initialize_wordcloud(self):
        return WordCloud(font_path=self.font_path,
                               width=2000,
                               height=1000,
                               background_color='white',
                               random_state=42,
                               collocations=False)
    
    def color_func(self, word, font_size, position, orientation, random_state=None,
                   **kwargs):
        hue_saturation = '42, 88%'

        current_emoji_probability = self.emoji_probability[word]
        if current_emoji_probability >= 0.10:
            opacity = 50
        else:
            opacity = 75 - current_emoji_probability/0.2 * 5
        return f"hsl({hue_saturation},{opacity}%)"

    def generate(self, text):
        emoji_frequencies = Counter(emojis.iter(text))
        total_count = sum(emoji_frequencies.values())
        
        self.emoji_probability = {emoji: count/total_count for emoji, count in emoji_frequencies.items()}
        wc = self.word_cloud.generate_from_frequencies(emoji_frequencies)
        
        plt.figure(figsize=(15,10))
        plt.imshow(wc.recolor(color_func=self.color_func, random_state=28))
        plt.axis("off")

tweets = pd.read_csv('/content/drive/MyDrive/cs620/tweets_n_spam.csv')
spams = tweets[tweets['spam'] == 1]
text = ' '.join(spams['text'])
emoji_cloud = EmojiCloud(font_path='emoji.ttf')
emoji_cloud.generate(text)
```

# **Discussion** üöß
-------

### Dynamics of emoji usage ‚úÖ

We observed\ the initial decrease in average number of emojis per tweet, which accounts for tweets that do not have emojis at all (i.e., we assume that pulling the data with Twitter API offers a representative sample of tweets), in 2015. This may be explained by the overal dynamics of daily tweets over the years, as other authors [found](https://blog.gdeltproject.org/visualizing-eight-years-of-twitters-evolution-2012-2019/) a decrease in daily tweeting rate after 2015:



tweets-per-day.png

### Life histories of emojis ‚úÖ

Individual life histories of emojis tells a little about the context of their usage. First, it is important to understand that emojis vary in their popularity (Zhao & Zeng 2017; see [the online tool to track emoji usage](http://emojitracker.com/)): e.g., the `FACE WITH TEARS OF JOY` üòÇ seems to be the most popular emoji both in our data and in the results of numerous studies of the topic (Zhao & Zeng 2017, Kerjiwal et al. 2021), followed by `HEAVY BLACK HEART` ‚ô•, `LOUDLY CRYING FACE` üò≠, `SMILING FACE WITH HEART-SHAPED EYES` üòç and so on. The distribution of individual emojis frequency has a high resemblance with species-abundance curve in ecology (Magurran 2004, Volkov et al. 2013), which implies that there are common mechanisms (or their absence, see Hubbell 2001) that govern the distribution of both species in biological communities and emojis in the Twitter ecosystem.

Second, it is unclear whether usage of the Twitter API provides even and representative sampling to ensure that the data on usage of less frequent emojis is accurate. Given that we obtained very few records of rare emojis (e.g., `GEMINI` ‚ôä, `JAPANESE ‚ÄúSERVICE CHARGE‚Äù BUTTON` üàÇÔ∏è, `SLOT MACHINE` üé∞, `NOTEBOOK WITH DECORATIVE COVER` üìî etc.), there are very few weeks in the data when we detected such emojis, although it is highly doubtful that they were not used on Twitter across these dates. It is unclear in which way sampling should be stratified to acurately represent the usage of emojis without over- or underestimating it, given the way the API queries are built and operate.

With that said, it must be accounted for the fact that the [data product](https://drive.google.com/file/d/1DlbilQRitV5DSTzDNsogv-E3EzUQejtA/view?usp=sharing) of this project which describes weekly frequencies of individual emojis in tweets might contain underestimated values for rare emojis.

The distribution of individual emojis' evolutionary success also resembles the classic biological evolution: there are many species with low success (i.e., such that their abundance neither fluctuates nor increases considerably and remains constantly low over time), and fewer and fewer emojis have a high success rate, implying that there among all 3 thousand emojis only few actually manage to become successful and are used disproportionately frequently.

To some extent, it could be explained by emoji's semantic meaning, as we observed that emojis with high positive score tend to have higher evolutionary success. Yet, we would expect then that emojis with high negative score are less popular, which is not the case: there seems to be no relationship between the negative score and succes rate, while the neutral score is actually negatively associated with the success rate. It implies that while positive emojis are used frequently, probably, due to their emotional intensity, negative emojis might have different connotations which makes it difficult to draw a consistent conclusion, whereas neutral emojis rarely have any emotional meaning, making them useless and ineffective memes.


## Emojis and tweet popularity ‚úÖ

There is a positive relationship between the tweet's likes count and number of emojis per word and the followers count, but a negative association with number of emojis per character and average negative score of emojis used. It implies that the users tend to like the tweets with an excess of negatively connotated emojis, but the tweets with reasonable counts of any emojis might actually get more feedback from the users.

In the same time, the retweet rate is positively affected by emoji richness, count of emojis per word and per character, followers count, and average negaive score of emojis used. It seems that the tweets containing some complains or rage may have more feedback in terms of retweeets (i.e., raising awareness about some important issue), while using emojis can actually help to draw attention to such an issue.

Finally, there is a negative effect of emoji richness on the reply rate. When all three tweet popularity metrics are combined together, it seems that the users perceive a saturated with emoji tweet as a signal to spread the information in this tweet (especially when there is some negative information within a tweet), but they would refrain from making comments when there is a variety of emojis used and would be less eager to like a tweet when there is a high emoji-to-character ratio.

*   Emojis usage used to be disproportionately popular immediately after they were introduced into Twitter in 2014, but eventually it stabilized at lower frequencies.
